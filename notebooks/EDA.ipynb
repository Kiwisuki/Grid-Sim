```python
%cd ..

import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import clear_output
import seaborn as sns
import warnings

from main import run_simulation


warnings.filterwarnings("ignore")
plt.rcParams['figure.dpi'] = 200

def get_peak_infected(ts):
    ts = pd.DataFrame(ts)
    return ts['Infected'].max() / ts.iloc[0].sum() * 100

peaks_a = []
peaks_b = []
for _ in range(300):
    _, ts_a, _, ts_b = run_simulation()
    peaks_a.append(get_peak_infected(ts_a))
    peaks_b.append(get_peak_infected(ts_b))

clear_output()

sns.distplot(peaks_a, label='No Quarantine')
sns.distplot(peaks_b, label='Quarantine')
plt.xlabel('Peak infected (%)')
plt.ylabel('Density')
plt.legend()
plt.show()
```
